# Section 1: Prey Species in the California Drylands

## The Data

This dataset contains information on arthropod prey species in California. It is also compiled with some environmental data, such as climate, elevation, and distance to nearest road. It is a synthesis of multiple datasets from sources such as the GBIF, USGS, and the U.S. Census, and it is currently being hosted by NCEAS. Prey data was acquired from Lortie et al. (2023)[ https://besjournals.onlinelibrary.wiley.com/doi/full/10.1002/2688-8319.12251]

Link to the archive: https://knb.ecoinformatics.org/view/doi%3A10.5063%2FF1VM49RH
